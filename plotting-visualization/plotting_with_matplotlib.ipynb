#### 4. Visualize data using a scatterplot.

```python
fig, ax = plt.subplots()

# Label plot
ax.set_title('Interesting Data')

# Set limits
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# Generate data
exponential_distribution = stats.expon()
x = exponential_distribution.rvs(10000)
y = exponential_distribution.rvs(10000)

# Plot
ax.scatter(x, y, color="blue")
```

##### What's wrong with the plot above? How is it fixed below?

```python
# Set figure size
#so the x axis has the same numbers and the y axis but is
#longer --- this is a scale issue 
fig, ax = plt.subplots(figsize=(5, 5))

# Label plot
ax.set_title('Interesting Data')

# Set limits
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# Generate data
#we are using scipy to make an exponential model 
exponential_distribution = stats.expon()
#we are asking for the generation of 10000 random points
x = exponential_distribution.rvs(10000)
y = exponential_distribution.rvs(10000)

# Plot with alpha transparency
ax.scatter(x, y, alpha=.05, color="purple")
```

#### 5. Visualize sequential data using a line plot.

```python
fig, ax = plt.subplots()

# Label plot
ax.set_title('Data Science Learning')

# Label axes
ax.set_xlabel('Week')
ax.set_ylabel('Mastery (%)')

# Set limits
ax.set_xlim(1, 15)
ax.set_ylim(0, 100)

x = np.arange(0, 16)
y = 40 + x**2/4

ax.plot(x, y, linestyle="dashed", color="green")
```

#### 6. Visualize data using a bar chart.

```python
import pandas as pd 

fig, ax = plt.subplots(figsize = (20,20))

# Label plot
ax.set_title('Cereal')

ax.set_xlabel('Cereal Name')
ax.set_ylabel('Rating')

# Generate data
cereal_df = pd.read_csv('cereal.csv').sort_values('rating')
x = cereal_df['name']
y = cereal_df['rating']

# Set limits
#ax.set_xlim(max(), min())
#ax.set_ylim(min(), max())

# Plot
ax.barh(x, y, color="blue")
```
